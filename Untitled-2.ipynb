```python
af_majority = df[df.rainfall == 1]
af_minority = df[df.rainfall == 0]
```

```python
print(af_majority.shape)
print(af_minority.shape)
```

```python
af_majority_downsampled = resample(af_majority, replace=False, n_samples=len(af_minority), random_state=42)
```

```python
af_majority_downsampled.shape
```

```python
af_majority_downsampled.head()
```

```python
df_downsampled = pd.concat([af_majority_downsampled, af_minority])
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
df_downsampled.head()
```

```python
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

```python
df_downsampled["rainfall"].value_counts()
```

```python
X = df_downsampled.drop(columns=["rainfall"])
y = df_downsampled["rainfall"]
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=92)
```

```python
rf_model = RandomForestClassifier(random_state=92)

param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_features": ["sqrt", "log2"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2)

grid_search_rf.fit(X_train, y_train)
```

Collecting workspace informationI'm sorry, I can't answer that question with what I currently know about your workspace.

```python
best_rf_model = grid_search_rf.best_estimator_

print("best parameters for Random Forest:", grid_search_rf.best_params_)
```

```python
cv_scores = cross_val_score(best_rf_model, X_train, y_train, cv=5)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))
```

```python
y_pred = best_rf_model.predict(X_test)

print("Test set Accuracy:", accuracy_score(y_test, y_pred))
print("Test set Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
